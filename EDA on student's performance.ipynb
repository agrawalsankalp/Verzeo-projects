### Grading Students on the Basis of percentage they achieved and analyzing the same with repect to different categories available 

```python
def dgrade(percentage):
    if percentage >= 90 and percentage <= 100:
        return 'Grade A'
    elif percentage >= 80 and percentage < 90:
        return 'Grade B'
    elif percentage >= 70 and percentage < 80:
        return 'Grade C'
    elif percentage >= 60 and percentage < 70:
        return 'Grade D'
    elif percentage >= 0 and percentage < 60:
        return 'Grade E'
    
sp['grades']=sp['percentage'].apply(dgrade)
```

```python
sp
```

```python
plt.figure(figsize=(15,6))
sp['grades'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

## observation
* Most students achieved Grade E followed by Grade C and Grade D

## The following plot will depict you how the Grades varies with the Race/Ethnicity , Parental Level of Education, Gender, lunch and Test perparation Course.

```python
plt.figure(figsize=(15,6))
plt.title('Grade Vs Ethnicity',fontsize = 18)
sns.countplot(x="grades", hue="race/ethnicity", data=sp)
plt.show()
plt.figure(figsize=(15,6))
plt.title('Grade Vs test preparation course',fontsize = 18)
sns.countplot(x="grades", hue="tpcourse", data=sp)
plt.show()
plt.figure(figsize=(15,6))
plt.title('Grade Vs lunch',fontsize = 18)
sns.countplot(x="grades", hue="lunch", data=sp)
plt.show()
plt.figure(figsize=(15,6))
plt.title('Grade Vs parental level of education',fontsize = 18)
sns.countplot(x="grades", hue="pdegree", data=sp)
plt.show()
plt.figure(figsize=(15,6))
plt.title('Grade Vs Gender',fontsize = 18)
sns.countplot(x="grades", hue="gender", data=sp)
plt.show()
```
